```python
# Split text into lines based on '|'
#lines = cl_text.split('Location Information')
lines = cl_text.split("\n")
# Create a list of dictionaries to hold each row of data
data = []
# Iterate over each line
for line in lines:
    data.append(line)

#trim the list
data = [item for item in data if item.strip()]
data
```

```python
# Replace "Location Type" with "Location Type" in each string
data = [re.sub(r'(?i)Location Information', '', string) for string in data]
data = [re.sub(r'(?i)Location Type', '| Location Type', string) for string in data]
data = [re.sub(r'(?i)Description|Descriptin|Descritpion|Descritpion|Descritpion|Descritpion|Descritption', '| Description', string) for string in data]
data = [re.sub(r'(?i)Sample size', '| Sample size', string) for string in data]

# Print the modified list
for string in data:
    print(string)
```

```python
pd.set_option('display.max_colwidth', None)

# Create DataFrame from the list
df = pd.DataFrame(data)
# Trim whitespace from all string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
# Display DataFrame
df.describe
```

```python
df.columns =['Location Information']
print(df.columns)
df.head()
```

```python

# Function to parse text and create key-value pairs
def parse_text(text):
    pairs = text.split('|')
    kv_pairs = {}
    for pair in pairs:
        parts = pair.split(':', 1)
        if len(parts) == 2:  # Check if split produced two parts
            key, value = parts
            kv_pairs[key.strip()] = value.strip()
        else:
            # Handle cases where the split doesn't produce two parts
            pass  # You can choose to ignore or handle such cases as needed
    return kv_pairs


# Apply the function to each row of the dataframe
df['Key_Value_Pairs'] = df['Location Information'].apply(parse_text)

print(df['Key_Value_Pairs'])
```
