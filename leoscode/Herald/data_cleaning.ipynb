```python
import pandas as pd
from IPython.display import display



# Load data from different worksheets in the same workbook
data1 = pd.read_excel('HH-cleaning.xlsx', sheet_name='data1')
data2 = pd.read_excel('HH-cleaning.xlsx', sheet_name='data2')

```

```python
# Merge the data
merged_data = pd.merge(data2, data1, on=['deviceid', 'end_mod_a','start_mod_a','hhid'], how='inner')
# Display the merged data
print(merged_data)
```

merged_data = pd.merge(data1, data2, on=None, how='left')  # Left join
merged_data = pd.merge(data1, data2, on=None, how='right')  # Right join
merged_data = pd.merge(data1, data2, on=None, how='outer')  # Full outer join

```python
display(merged_data)
```

```python
merged_data.to_excel('merged_data.xlsx', index=False)
```
