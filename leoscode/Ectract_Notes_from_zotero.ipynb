```python

def define_patterns():
    html_tag_pattern = re.compile(r'<.*?>')
    nbsp_pattern = re.compile(r'&nbsp;')
    amp_pattern = re.compile(r'&amp;')
    nn_pattern = re.compile(r'\n;')
    notes_pattern = re.compile(r'Notes;')
    location_information_pattern = re.compile(r'Location Information|Location information ')
    description_pattern = re.compile(r'Description|Descriptin|Descritpion|Descritpion|Descritpion|Descritpion|Descritption|description ')
    sample_size_pattern = re.compile(r'Sample size|sample size|Sample Size')
    location_type_pattern = re.compile(r'Location type|Location Type|location type')
    location_name_pattern = re.compile(r'Location name|location name|location Name|Location Names')
    return {
        'html_tag_pattern': html_tag_pattern,
        'nbsp_pattern': nbsp_pattern,
        'amp_pattern': amp_pattern,
        'nn_pattern': nn_pattern,
        'notes_pattern': notes_pattern,
        'location_information_pattern': location_information_pattern,
        'description_pattern': description_pattern,
        'sample_size_pattern': sample_size_pattern,
        'location_type_pattern': location_type_pattern,
        'location_name_pattern': location_name_pattern
    }

def remove_html_tags(df, patterns):
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['html_tag_pattern'], '', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['nbsp_pattern'], '', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['amp_pattern'], '', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['nn_pattern'], '', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['notes_pattern'], '', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['location_information_pattern'], '', x))
    return df

def replace_labels(df, patterns):
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['description_pattern'], 'Description', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['sample_size_pattern'], 'Sample Size', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['location_type_pattern'], 'Location Type', x))
    df['Notes'] = df['Notes'].apply(lambda x: re.sub(patterns['location_name_pattern'], 'Location Name', x))
    return df

def remove_spaces_before_labels(df):
    df['Notes'] = df['Notes'].str.replace(r'\s+Location Name', 'Location Name')
    df['Notes'] = df['Notes'].str.replace(r'Location Type', '| Location Type')
    df['Notes'] = df['Notes'].str.replace(r'Description', '| Description')
    df['Notes'] = df['Notes'].str.replace(r'Sample Size', '| Sample Size')
    return df

#def main():
patterns = define_patterns()
df = remove_html_tags(df, patterns)
df = replace_labels(df, patterns)
df = remove_spaces_before_labels(df)
print(df)

#if __name__ == "__main__":
 #   main()
```

```python
# Function to parse text and create key-value pairs
def parse_text(text):
    pairs = text.split('|')
    kv_pairs = {}
    for pair in pairs:
        parts = pair.split(':', 1)
        if len(parts) == 2:  # Check if split produced two parts
            key, value = parts
            kv_pairs[key.strip()] = value.strip()
        else:
            # Handle cases where the split doesn't produce two parts
            pass  # You can choose to ignore or handle such cases as needed
    return kv_pairs


# Apply the function to each row of the dataframe
df['Key_Value_Pairs'] = df['Notes'].apply(parse_text)

print(df['Key_Value_Pairs'])
```

```python
# Define the regular expression pattern with case-insensitive flag

health_facility_pattern = re.compile(r'Health faciilities|Health facility, districts, Community, Community|Health facilityt|Health facility, Research Institute|Health facilityt|12 Health facility|4 Health facility|Health facility/ City|Health facilityt|City, Health facility|Health facility \(antenatal clinics\)|10 Public Antenatal Clinics \(Health facility\)|Health facility / District|Health facility and Community|Communityies|Health facilityt|Health facility Catchmen|District, Health facility|Health faciilities|Health facility in Johannesburg.|Study Cluster/ Health facility|Health facility, districts, Community, Community|Health facilities|Health facilities|health facilities|health facility|Health facilities|Health Facility|Health facilities|Health facilities|health facilities|Healty Facility|health facility|Health Facilities|12 health facilities|4 Health facilities|12 health facilities|10 Public Antenatal Clinics \(health facilities\)|Health facility  District|Community Health facility|Health facility and community|Health facility|Villages / Health facility|Health facility and Communities|Health centers|Public Health facility|City / Health Facility|4 Health facilities|Healthy Facility|Health facilities|Health facilities|Health facilities|Health facilities|Health facilities')
Community_pattern = re.compile(r'community/ City|Communityy|Community in Zambia|Rural community|Community and Health facility|rural areas|urban areas|Health facility, Township|Health facility and communit|Health facility and communities|Communities|Rural communities|Community and Health Facilities in Zambia|Community sites|Communities/Global Network sites / study cluster|Rural Community')
research_center_pattern = re.compile(r'Reaserch cente|research centers|Research Instituter|Research Instituter, Research Institute')

# Remove HTML tags from the 'notes' column
df['Location Type'] = df['Location Type'].astype(str)
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(health_facility_pattern, 'Health facility', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(Community_pattern, 'Community', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(research_center_pattern, 'Research Institute', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(r'Community/Global Network sites / study cluster|Community in Zambia|Communityy', 'Community', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(r'Community/ City|Communityies', 'Community', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(r'City / Health facility|City, Health facility|Health facility, Research Institute|Health facility \(antenatal clinics\)|Health facility in Johannesburg.|Health facility and Community|Community Health facility|Health facility, districts, Community, Community|Health facility Catchment|Public Health facility|Villages / Health facility|Study Cluster/ Health facility', 'Health facility', x))
df['Location Type'] = df['Location Type'].apply(lambda x: re.sub(r'Research Instituter', 'Research Institute', x))
```
