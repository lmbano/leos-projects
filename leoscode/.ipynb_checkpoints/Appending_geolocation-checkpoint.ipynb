# Kenya Villages coodinates extraction

```python
import geopy
from geopy.geocoders import Nominatim
import pandas as pd

# Load Kenya Presice dataset for geocodes extraction
df_kenya = pd.read_excel('Kenya_Villages.xlsx')

# Drop columns with NAN values in the village column
df_kenya = df_kenya.dropna(subset=['village'])

# Drop duplicate village names in the village column
df_kenya = df_kenya.drop_duplicates(subset='village')

# Create a geolocator object
geolocator = Nominatim(user_agent="my_app")

# Define a function to get coordinates
def get_coordinates(location_name):
    location = geolocator.geocode(location_name, timeout=10)
    if location and "Kenya" in location.address and "Kilifi" in location.address:  # Filter for Kilifi county
        return (location.latitude, location.longitude)
    else:
        return (None, None)

# Apply the function to the 'village' column
df_kenya[['Latitude', 'Longitude']] = df_kenya['village'].apply(get_coordinates).apply(pd.Series)
```


# Extracting Geo-coodinates from 


```python
import geopy
from geopy.geocoders import Nominatim
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

df_kenya = pd.read_excel('Kenya_Villages.xlsx')
df_kenya = df_kenya.dropna(subset=['village'])
df_kenya = df_kenya.drop_duplicates(subset='village')

geolocator = Nominatim(user_agent="my_app", timeout=10)

def get_coordinates(location_name):
    location = geolocator.geocode(location_name, language="en")
    if location:
        return (location.latitude, location.longitude)
    else:
        return (None, None)

def geocode_villages(village_names):
    with ThreadPoolExecutor(max_workers=5) as executor:
        coordinates = list(executor.map(get_coordinates, village_names))
    return coordinates

village_names = df_kenya['village'].tolist()
coordinates = geocode_villages(village_names)

batches = [df_kenya[i:i+100] for i in range(0, len(df_kenya), 100)]

# perform some translations on each batch
translated_batches = []
for batch in batches:
    # perform your translations here, for example:
    batch[['Latitude', 'Longitude']] = pd.DataFrame(coordinates, index=village_names)
    translated_batches.append(batch)

# remerge the batches
merged_df = pd.concat(translated_batches, ignore_index=True)
```

# Using Open Case Geo Coder API

```python
import pandas as pd
import requests

df_kenya = pd.read_excel('Kenya_Villages.xlsx')

def geocode_villages(village_names):
    api_key = "YOUR_OPENCAGE_GEOCODER_API_KEY"
    base_url = "https://api.opencagedata.com/geocode/v1/json"
    coordinates = []
    for village in village_names:
        params = {
            "q": village,
            "key": api_key
        }
        response = requests.get(base_url, params=params)
        data = response.json()
        if data["results"]:
            coordinates.append((data["results"][0]["geometry"]["lat"], data["results"][0]["geometry"]["lng"]))
        else:
            coordinates.append((None, None))
    return coordinates

village_names = df_kenya['village'].tolist()
coordinates = geocode_villages(village_names)

df_kenya[['Latitude', 'Longitude']] = pd.DataFrame(coordinates, index=village_names)
```

# Using MapQuest API

```python
import pandas as pd
import requests

df_kenya = pd.read_excel('Kenya_Villages.xlsx')

def geocode_villages(village_names):
    api_key = "YOUR_MAPQUEST_API_KEY"
    base_url = "http://www.mapquestapi.com/geocoding/v1/address"
    coordinates = []
    for village in village_names:
        params = {
            "key": api_key,
            "location": village
        }
        response = requests.get(base_url, params=params)
        data = response.json()
        if data["results"][0]["locations"]:
            coordinates.append((data["results"][0]["locations"][0]["latLng"]["lat"], data["results"][0]["locations"][0]["latLng"]["lng"]))
        else:
            coordinates.append((None, None))
    return coordinates

village_names = df_kenya['village'].tolist()
coordinates = geocode_villages(village_names)

df_kenya[['Latitude', 'Longitude']] = pd.DataFrame(coordinates, index=village_names)
```
