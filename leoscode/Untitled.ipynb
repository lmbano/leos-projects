```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a new figure
fig, ax = plt.subplots(figsize=(12, 8))

# Define the flowchart elements
elements = [
    {"label": "Load data", "pos": (0.5, 0.9)},
    {"label": "Feature engineering", "pos": (0.5, 0.75)},
    {"label": "Encoeding using One-hot", "pos": (0.5, 0.6)},
    {"label": "emberding", "pos": (0.5, 0.45)},
    {"label": "Plit data into X & Y sets", "pos": (0.2, 0.3)},
    {"label": "Create model, train/ fit", "pos": (0.8, 0.3)},
    {"label": "Evealuate results & Finetune", "pos": (0.5, 0.15)},
    {"label": "Recommendations", "pos": (0.5, 0)}
]


# Draw the elements and arrows
for elem in elements:
    ax.text(elem["pos"][0], elem["pos"][1], elem["label"],
            ha='center', va='center', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightblue'))

# Draw arrows
for i in range(len(elements) - 1):
    ax.annotate("", xy=elements[i + 1]["pos"], xycoords='data',
                xytext=elements[i]["pos"], textcoords='data',
                arrowprops=dict(arrowstyle="->", lw=1.5))

# Set limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')

# Title
plt.title("Flowchart of Neural Network Embedding based recommender system", fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def plot_flowchart():
    fig, ax = plt.subplots(figsize=(12, 8))

    # Set title
    ax.set_title('Course Similarity-Based Recommender System Flowchart', fontsize=18, fontweight='bold')

    # Define boxes
    start = mpatches.Rectangle((0.1, 0.8), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    load_data = mpatches.Rectangle((0.4, 0.8), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    preprocess_data = mpatches.Rectangle((0.7, 0.8), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    create_bow = mpatches.Rectangle((0.1, 0.5), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    calculate_similarity = mpatches.Rectangle((0.4, 0.5), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    filter_courses = mpatches.Rectangle((0.7, 0.5), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    rank_courses = mpatches.Rectangle((0.1, 0.2), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    recommend_courses = mpatches.Rectangle((0.4, 0.2), 0.2, 0.1, edgecolor='black', facecolor='lightblue')
    end = mpatches.Rectangle((0.7, 0.2), 0.2, 0.1, edgecolor='black', facecolor='lightblue')

    # Add boxes to axes
    ax.add_patch(start)
    ax.add_patch(load_data)
    ax.add_patch(preprocess_data)
    ax.add_patch(create_bow)
    ax.add_patch(calculate_similarity)
    ax.add_patch(filter_courses)
    ax.add_patch(rank_courses)
    ax.add_patch(recommend_courses)
    ax.add_patch(end)

    # Add text to boxes
    ax.text(0.15, 0.85, 'Start', ha='center')
    ax.text(0.45, 0.85, 'Load Course Data', ha='center')
    ax.text(0.75, 0.85, 'Preprocess Data', ha='center')
    ax.text(0.15, 0.55, 'Create BoW', ha='center')
    ax.text(0.45, 0.55, 'Calculate Similarity', ha='center')
    ax.text(0.75, 0.55, 'Filter Courses', ha='center')
    ax.text(0.15, 0.25, 'Rank Courses', ha='center')
    ax.text(0.45, 0.25, 'Recommend Courses', ha='center')
    ax.text(0.75, 0.25, 'End', ha='center')

    # Add arrows
    ax.arrow(0.3, 0.8, 0.1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')
    ax.arrow(0.6, 0.8, 0.1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')
    ax.arrow(0.3, 0.5, 0.1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')
    ax.arrow(0.6, 0.5, 0.1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')
    ax.arrow(0.3, 0.2, 0.1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')

    # Set limits
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    # Hide axes
    ax.axis('off')

    plt.show()

plot_flowchart()
```
